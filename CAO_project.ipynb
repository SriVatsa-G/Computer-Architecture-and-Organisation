```python
df.drop(['JobID','JobStructure','JobStructureParams','UsedNetwork','UsedLocalDiskSpace','UsedResources','ReqPlatform','ReqNetwork',
          'ReqLocalDiskSpace','ReqResources','VOID','ProjectID'],axis=1,inplace=True)
```

```python
df['QueueNO'] = df['QueueID'].rank(method='dense', ascending=False).astype(int)
df['GroupNO'] = df['GroupID'].rank(method='dense', ascending=False).astype(int)
df['ExecutableNO'] = df['ExecutableID'].rank(method='dense', ascending=False).astype(int)
df['OrigSiteNO'] = df['OrigSiteID'].rank(method='dense', ascending=False).astype(int)
df['LastRunSiteNO'] = df['LastRunSiteID'].rank(method='dense', ascending=False).astype(int)
df.head()
```

```python
df['USERNO'] = df['UserID'].rank(method='dense', ascending=False).astype(int)
df.head()
```

```python
df.drop(['NProc','UserID','QueueID','GroupID','ExecutableID','OrigSiteID','LastRunSiteID'],axis=1,inplace=True)
```

```python
df["ReqNProcs"].unique()
```

```python
data = df
```

**MODEL**

```python
# Split your data
X = data.drop('ReqNProcs', axis=1)
y = data['ReqNProcs']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale your data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

ridge_regressor = Ridge(random_state=42)
ridge_regressor.fit(X_train_scaled, y_train)
y_pred_ridge = ridge_regressor.predict(X_test_scaled)

rmse_ridge = mean_squared_error(y_test, y_pred_ridge, squared=False)
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f'Ridge Regression:\nRMSE: {rmse_ridge}\nR^2: {r2_ridge}\n')
```

```python
from sklearn.linear_model import Lasso

lasso_regressor = Lasso(random_state=42)
lasso_regressor.fit(X_train_scaled, y_train)
y_pred_lasso = lasso_regressor.predict(X_test_scaled)

rmse_lasso = mean_squared_error(y_test, y_pred_lasso, squared=False)
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f'Lasso Regression:\nRMSE: {rmse_lasso}\nR^2: {r2_lasso}\n')
```

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree_regressor = DecisionTreeRegressor(random_state=42)
decision_tree_regressor.fit(X_train_scaled, y_train)
y_pred_dt = decision_tree_regressor.predict(X_test_scaled)

rmse_dt = mean_squared_error(y_test, y_pred_dt, squared=False)
r2_dt = r2_score(y_test, y_pred_dt)

print(f'Decision Tree Regression:\nRMSE: {rmse_dt}\nR^2: {r2_dt}\n')
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_regressor = RandomForestRegressor(random_state=42)
random_forest_regressor.fit(X_train_scaled, y_train)
y_pred_rf = random_forest_regressor.predict(X_test_scaled)

rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)

print(f'Random Forest Regression:\nRMSE: {rmse_rf}\nR^2: {r2_rf}\n')
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gradient_boosting_regressor = GradientBoostingRegressor(random_state=42)
gradient_boosting_regressor.fit(X_train_scaled, y_train)
y_pred_gb = gradient_boosting_regressor.predict(X_test_scaled)

rmse_gb = mean_squared_error(y_test, y_pred_gb, squared=False)
r2_gb = r2_score(y_test, y_pred_gb)

print(f'Gradient Boosting Regression:\nRMSE: {rmse_gb}\nR^2: {r2_gb}\n')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error, r2_score

# Define the model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)  # Output layer for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Fit the model
history = model.fit(X_train_scaled, y_train, validation_split=0.2, epochs=100, verbose=1)

# Predict on the test data
y_pred_nn = model.predict(X_test_scaled)

# Since the output will be in a 2D array (n_samples, 1), we flatten it to a 1D array for comparison
y_pred_nn = y_pred_nn.flatten()

# Calculate RMSE and R-squared
rmse_nn = mean_squared_error(y_test, y_pred_nn, squared=False)
r2_nn = r2_score(y_test, y_pred_nn)

print(f'Neural Network Regression:\nRMSE: {rmse_nn}\nR^2: {r2_nn}\n')
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

# Random Forest model
rf = RandomForestRegressor()

# Adjusted parameters to tune
param_dist = {
    'n_estimators': [100, 150, 200],  # reduced upper limit
    'max_depth': [10, 20, 30, 40, None],  # added None for unlimited depth
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Randomized search
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=50, cv=3, verbose=2, random_state=42, n_jobs=-1)

# Fit the random search model
rf_random.fit(X_train_scaled, y_train)

# Making predictions
rf_predictions = rf_random.predict(X_test_scaled)

mse = mean_squared_error(y_test, rf_predictions)
r2 = r2_score(y_test, rf_predictions)

print("MSE:", mse)
print("R-squared:", r2)
# Best parameters
print("Best Parameters: ", rf_random.best_params_)
```
